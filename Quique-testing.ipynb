```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
from us_state_abbrev import us_state_abbrev
from folium.plugins import HeatMap
```

### Trying out MarkerCluster

```python
df_breweries = pd.read_csv('clean_data/breweries_latlng.csv')
```

```python
breweries_state_map = folium.Map(location=[48, -102], zoom_start=3)
marker_cluster = MarkerCluster().add_to(breweries_state_map)

for each in df_breweries.iterrows():
    folium.Marker(
        location = [each[1]['latitude'],each[1]['longitude']]).add_to(marker_cluster)
```

### Trying out HeatMap

```python
data = df_breweries.as_matrix(['latitude', 'longitude']).tolist()
```

```python
m = folium.Map(location=[48, -102],tiles='stamentoner', zoom_start=3)

HeatMap(data,radius=15).add_to(m)
m
```
